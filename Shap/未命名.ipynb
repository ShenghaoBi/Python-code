```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import shap

# 
np.random.seed(0) # 
n_samples = 1000
X = np.random.rand(n_samples, 6) # X1X6
Y = np.random.rand(n_samples) # Y

# DataFrame
data = pd.DataFrame(X, columns=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])
data['Y'] = Y

# 
rf_model = RandomForestRegressor(n_estimators=100, random_state=0)
rf_model.fit(X, Y) # 

# SHAP
explainer = shap.Explainer(rf_model, X) 
shap_values = explainer.shap_values(X) 

# SHAP
shap.summary_plot(shap_values, X, feature_names=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])

```

```python
# SHAP Force Plot
shap.initjs()
sample_index = 0 # 
shap.force_plot(explainer.expected_value, shap_values[sample_index], features=X[sample_index], feature_names=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])

# SHAP Scatter Plot
shap.summary_plot(shap_values, X, plot_type='scatter')

# SHAP Beeswarm Plot
shap.summary_plot(shap_values, X, plot_type='violin')

```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import shap
import matplotlib.pyplot as plt

# 
np.random.seed(0)
n_samples = 1000
X = np.random.rand(n_samples, 6)
Y = np.random.rand(n_samples)

# DataFrame
data = pd.DataFrame(X, columns=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])
data['Y'] = Y

# 
rf_model = RandomForestRegressor(n_estimators=100, random_state=0)
rf_model.fit(X, Y)

# SHAP
explainer = shap.Explainer(rf_model, X)
shap_values = explainer.shap_values(X)

# SHAP Summary Plot
shap.summary_plot(shap_values, X, feature_names=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])

# SHAP Force Plot
sample_index = 0
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values[sample_index], features=X[sample_index], feature_names=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])

# SHAP Scatter Plot
shap.summary_plot(shap_values, X, plot_type='scatter')

# SHAP Beeswarm Plot
shap.summary_plot(shap_values, X, plot_type='violin')

plt.show()

```
